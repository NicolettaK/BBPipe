### Check whether amplitude of output fg maps is consistent with data
- We get fg at 2.3 and 353 GHz, to compare with S-PASS and HFI data

```python
cmb, dust, synch = mk_mock_data.get_sky_sim_gaus(256, 0., 1., freqs=[2.3, 353.], 
                                                 thermo_units=True)
```

- We check fg amplitude on the inner region of SO sky mask, covering about 10% of the sky

```python
hits = hp.read_map('/Users/niki/Workspace/SimonsObs/SO_Noise_Public/mask_SAT_apodized.fits')
mask = np.copy(hits)
mask[mask>0.4] = 1
mask[mask!=1] = 0
mask = hp.ud_grade(mask, 256)
print(len(mask[mask==1])/float(len(mask)))
```

- load S-PASS and HFI-353 maps, that we use to check amplitudes of synchrotron and dust respectively

```python
spass = np.array(hp.read_map(
    '/Users/niki/Workspace/SimonsObs/maps/Spass_Nov_QU_original_256.fits', (0,1,2)))*1e6
hfi = np.array(hp.read_map(
    '/Users/niki/Workspace/SimonsObs/maps/TQU_HFI_SkyMap_353_256_DX11d_full.fits', (0,1,2)))*1e6
```

- smooth HFI and gaussian dust map at one degree (as otherwise hfi data are noisy)

```python
hfi_1deg = hp.smoothing(hfi, np.radians(1.))
dust353 = dust[1]
dust353_1deg = hp.smoothing(dust353, np.radians(1.))
synch2p3 = synch[0]
```

- compare spass with gaussian synchrotron

```python
plt.figure(figsize=(10, 6))
hp.mollview(spass[1]*mask, min=-5e4, max=5e4, title='S-PASS Q', unit = 'muK', sub=121)
hp.mollview(synch2p3[1]*mask, min=-5e4, max=5e4, title='gauss synch Q', unit='muK', sub=122)
```

- compare HFI-353 with gaussian dust

```python
plt.figure(figsize=(10, 6))
hp.mollview(hfi_1deg[1]*mask, min=-100, max=100, title='HFI Q', unit = 'muK', sub=121)
hp.mollview(dust353_1deg[1]*mask, min=-100, max=100, title='gauss dust Q', unit='muK', sub=122)
```
